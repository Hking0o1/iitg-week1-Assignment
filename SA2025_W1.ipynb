# Identify Car Names with Multiple Model Years
Group the data by 'name' and count unique 'model_year' values to find car names appearing in more than one year.

```python
name_year_counts = df.groupby('name')['model_year'].nunique()
multi_year_names = name_year_counts[name_year_counts > 1].index.tolist()
print(f"Number of car names appearing in more than one model_year: {len(multi_year_names)}")
```

# Compute Standard Deviation of MPG for Each Car Name
For each car name with multiple years, calculate the standard deviation of 'mpg' across those years.

```python
# Filter data for car names with multiple years
multi_year_cars = df[df['name'].isin(multi_year_names)]

# Group by name and compute std and count for mpg
mpg_stats = (
    multi_year_cars.groupby('name')
    .agg(
        mpg_std=('mpg', 'std'),
        appearances=('mpg', 'count'),
        avg_mpg=('mpg', 'mean')
    )
    .reset_index()
)
mpg_stats = mpg_stats.dropna(subset=['mpg_std'])
mpg_stats.head()
```

# Filter Consistent Car Models (std &lt; 1.0, appearances ≥ 2)
Filter car names with at least 2 appearances and standard deviation of mpg less than 1.0.

```python
consistent_cars = mpg_stats[(mpg_stats['mpg_std'] < 1.0) & (mpg_stats['appearances'] >= 2)]
consistent_cars = consistent_cars.sort_values('mpg_std')
consistent_cars
```

# Find Car(s) with Lowest MPG Variation
Identify the car(s) among the filtered set with the lowest standard deviation in mpg.

```python
if not consistent_cars.empty:
    min_std = consistent_cars['mpg_std'].min()
    most_consistent = consistent_cars[consistent_cars['mpg_std'] == min_std]
else:
    most_consistent = pd.DataFrame()
most_consistent
```

# Report Model Name(s), Number of Appearances, and Average MPG
Display the model name(s), number of appearances, and average mpg for the most consistent car(s).

```python
if not most_consistent.empty:
    display_cols = ['name', 'appearances', 'avg_mpg']
    print("Most consistent car model(s):")
    display(most_consistent[display_cols])
else:
    print("No car models found with std(mpg) < 1.0 and at least 2 appearances.")
    
```

# Bonus: Sort Results by Number of Appearances and MPG
Sort the final results by number of appearances (descending), then by average mpg (descending).

```python
if not consistent_cars.empty:
    sorted_consistent = consistent_cars.sort_values(
        by=['appearances', 'avg_mpg'], ascending=[False, False]
    )
    display_cols = ['name', 'appearances', 'avg_mpg', 'mpg_std']
    print("Consistent car models sorted by appearances and average mpg:")
    display(sorted_consistent[display_cols])
else:
    print("No consistent car models to sort.")
```
